## Función para descargar y guardar los datasets
En esta función se realiza la operación de descargar y actualizar de forma incremental cada uno de los datasets definidos en la celda anterior. Si el dataset no existe, entonces crea uno con el nombre asignado, pero si ya existe entonces toma la última fecha registrada, la reescribe y luego ingresa nuevas filas con las fechas actualizadas teniendo como paso la ventana de tiempo definida para cada dataset.

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
import os
import pandas as pd


def actualizar_dataset_xm(dataset_row, objetoAPI):
    """
    Descarga y actualiza un dataset de XM de forma incremental
    según la configuración definida en dataset_row.
    """

    if not dataset_row.get("activo", True):
        print(f"[{dataset_row['dataset_id']}] Dataset inactivo. Se omite.")
        return None

    fecha_inicio = dataset_row["fecha_inicio"]
    fecha_fin = dataset_row["fecha_fin"]
    ACTUALIZAR_DESDE_CERO = dataset_row["ACTUALIZAR_DESDE_CERO"]

    ruta_base = dataset_row["ruta_base"]
    nombre_archivo = dataset_row["nombre_archivo"]
    ruta_archivo = os.path.join(ruta_base, nombre_archivo)

    metric_id = dataset_row["MetricID"]
    entity = dataset_row["Entity"]

    os.makedirs(ruta_base, exist_ok=True)

    # =========================
    # FUNCIÓN DE DESCARGA
    # =========================

    def descargar_xm(fecha_ini, fecha_fin):
        return objetoAPI.request_data(
            metric_id,
            entity,
            fecha_ini,
            fecha_fin
        )

    # =========================
    # DEFINIR FECHA DE INICIO EFECTIVA
    # =========================

    # =========================
    # DEFINIR FECHA DE INICIO EFECTIVA
    # =========================

    # =========================
# DEFINIR FECHA DE INICIO EFECTIVA
# =========================

    df_total = None   # inicialización segura

    if os.path.exists(ruta_archivo) and not ACTUALIZAR_DESDE_CERO:

        df_existente = pd.read_csv(ruta_archivo)

        df_existente["Date"] = pd.to_datetime(
            df_existente["Date"],
            format="mixed",
            errors="coerce"
        )

        df_existente = df_existente[df_existente["Date"].notna()]

        # Última fecha existente
        fecha_max = df_existente["Date"].max()

        # 🔹 empezar descarga después de la última fecha
        fecha_actual = fecha_max + pd.Timedelta(days=1)

        # 🔹 mantener histórico
        df_total = df_existente.copy()

        print(
            f"[{dataset_row['dataset_id']}] "
            f"Reanudando desde {fecha_actual.date()} "
            f"(última fecha guardada {fecha_max.date()})"
        )

    else:

        fecha_actual = fecha_inicio

        if ACTUALIZAR_DESDE_CERO and os.path.exists(ruta_archivo):

            print(f"[{dataset_row['dataset_id']}] Reescribiendo dataset desde cero.")

            # si se reescribe, no se carga histórico
            df_total = None

        else:

            print(f"[{dataset_row['dataset_id']}] Iniciando descarga completa.")
            df_total = None

    # =========================
    # DESCARGA ITERATIVA (X MESES)
    # =========================

    paso_mes = relativedelta(months=dataset_row["paso_meses"])

    while fecha_actual < fecha_fin:

        fecha_siguiente = min(
            fecha_actual + paso_mes,
            fecha_fin
        )

        print(
            f"[{dataset_row['dataset_id']}] "
            f"Descargando {fecha_actual.date()} → {fecha_siguiente.date()}"
        )

        df_nuevo = descargar_xm(fecha_actual, fecha_siguiente)

        df_nuevo["Date"] = pd.to_datetime(
            df_nuevo["Date"],
            format="mixed",
            errors="coerce"
        )

        df_nuevo = df_nuevo[df_nuevo["Date"].notna()]

        if df_total is None:
            df_total = df_nuevo
        else:
            df_total = (
                pd.concat([df_total, df_nuevo], ignore_index=True)
                .drop_duplicates()
                .sort_values("Date")
            )

        df_total.to_csv(ruta_archivo, index=False)

        print(
            f"[{dataset_row['dataset_id']}] "
            f"Registros acumulados: {len(df_total):,}"
        )

        fecha_actual = fecha_siguiente

   

    print(f"[{dataset_row['dataset_id']}] Dataset actualizado correctamente.")

    # =========================
    # DEVOLVER FECHA DE ACTUALIZACIÓN
    # =========================

    return datetime.now()
```

## PIPELINE de actualización
En esta celda se realiza la conexión con XM, y luego para cada fila de la tabla de datasets se realiza la actualización de los datos. Es importante tener en cuenta que esta se peude interrumpir en cualquier momento por demoras para acceder a la API ya que si el requerimineto tarda más de 90 segundos, desde XM lo eliminan, sin embargo, el código guarda el archivo hasta donde se logró ejecutar y luego al ejecutar de nuevo continúa desde donde había quedado, también se recomienda que si para un dataset se corta muchas veces el acceso poner una ventana de tiempo "paso_meses" más pequeña y volver a ejecutar el código.

```python
from pydataxm import pydataxm

# Conexión API XM
objetoAPI = pydataxm.ReadDB()

# Pipeline maestro
for idx, dataset_row in df_datasets.iterrows():
    fecha_actualizacion = actualizar_dataset_xm(
        dataset_row,
        objetoAPI
    )
```
